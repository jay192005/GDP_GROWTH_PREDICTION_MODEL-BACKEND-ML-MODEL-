```python
import pandas as pd

df_raw = pd.read_excel(r'final refinement.xlsx', engine='openpyxl')
df_clean = pd.read_csv(r'fully_corrected_data.csv')

df_raw = df_raw.sort_values(by=['Country', 'Year'])

df_years_adjusted = df_raw.groupby('Country', group_keys=False).apply(lambda x: x.iloc[1:])

df_clean['Year'] = df_years_adjusted['Year'].astype(int).values

cols = ['Country', 'Year'] + [c for c in df_clean.columns if c not in ['Country', 'Year']]
df_final = df_clean[cols]

df_final.to_csv('final_data_with_year.csv', index=False)

print(df_final.head())
```
